```python
def add_elm(d,q,w,e,r,t,y,u,i,o):
    d['Rank'] = q
    d['Team'] = w
    d['GP'] = e
    d['Won'] = r
    d['Lost'] = t
    d['Tied']=y
    d['Pct.']=u
    d['First NFL Season']=i
    d['Division']=o
    return d
```

```python
l = []

d1 = dict()
d2 = dict()
d3 = dict()
d4 = dict()
d5 = dict()


add_elm(d1,1,'Dallas Cowboys',914,520,388,6,.572,1960,'NFC East')
add_elm(d2,2,'Green Bay Packers',1368,756,574,38,.567,1921,'NFC North')
add_elm(d3,3,'New England Patriots',916,512,395,9,.564,1960,'AFC East')
add_elm(d4,4,'Chicago Bears',1402,769,591,42,.563,1920,'NFC North')
add_elm(d5,5,'Baltimore Ravens',384,214,169,1,.559,1996,'AFC North')

l.append(d1)
l.append(d2)
l.append(d3)
l.append(d4)
l.append(d5)


for i in range(len(l)):
    print(l[i])
```

```python
nfl_frame = DataFrame(l)
```
