```python
import pandas as pd
import numpy as np

from numpy.random import randn
from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("darkgrid")
```

```python
flight_dframe = sns.load_dataset('flights')
```

```python
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot('month','year','passengers')
```

```python
flight_dframe
```

```python
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
sns.heatmap(flight_dframe,linewidths=1,cmap='Reds')
```

```python
sns.heatmap(flight_dframe,linewidths=1,cmap='Reds',annot=True,fmt='d',annot_kws={"size": 12.5})
```

```python
sns.heatmap(flight_dframe,linewidths=1,center=flight_dframe.loc['January',1955])
```

```python
f, (axis1,axis2) = plt.subplots(2,1)

yearly_flights = flight_dframe.sum()

years=pd.Series(yearly_flights.index.values)
years=pd.DataFrame(years)

flights=pd.Series(yearly_flights.values)
flights=pd.DataFrame(flights)

year_dframe=pd.concat((years,flights),axis=1)
year_dframe.columns=['Year','Flights']

sns.barplot('Year',y='Flights',data=year_dframe,ax=axis1)

sns.heatmap(flight_dframe,cmap='RdBu_r',linewidths=1,ax=axis2,cbar_kws={'orientation':'horizontal'})

plt.rcParams['figure.figsize'] = (10, 5)
plt.tight_layout()
```

```python
sns.clustermap(flight_dframe,linewidths=1,cmap='RdBu_r')
```

```python
sns.clustermap(flight_dframe,linewidths=1,cmap='RdBu_r',col_cluster=False)
```

```python
sns.clustermap(flight_dframe,linewidths=1,cmap='Reds',standard_scale=1)
```

```python
sns.clustermap(flight_dframe,linewidths=1,cmap='Reds',standard_scale=0)
```

```python
sns.clustermap(flight_dframe,linewidths=1,cmap='RdBu_r',z_score=1)
```
