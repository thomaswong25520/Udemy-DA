```python
import pandas as pd
import numpy as np

from numpy.random import randn
from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("darkgrid")
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',order=4,markers='o',
          scatter_kws={'color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',fit_reg=False,markers='.')
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=0.1)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','.'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day',markers='.')
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','tip_pect',tips)
```

```python
pd.set_option('display.width', 1000)
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)

sns.regplot('total_bill','tip_pect',tips,ax=axis1)

sns.violinplot(x=tips['size'],y=tips['tip_pect'],data=tips,palette='Reds_r',ax=axis2)
```
