```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
sns.jointplot(data1,data2)
```

```python
from warnings import filterwarnings
filterwarnings('ignore')

sns.set(style="darkgrid", color_codes=True)
sns.jointplot(data1,data2, kind='reg', height=8).annotate(stats.pearsonr)
```

```python
from warnings import filterwarnings
filterwarnings('ignore')

sns.set(style="darkgrid", color_codes=True)
sns.jointplot(data1,data2, kind='hex', height=8).annotate(stats.pearsonr)
```
